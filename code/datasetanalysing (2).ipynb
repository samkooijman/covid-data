```python
#Denmark Vaccination

vacc_den = pd.read_csv("data/teamproject/vacc_den.csv",  encoding= 'unicode_escape', delimiter=",")
vacc_den = vacc_den.rename(columns=str.lower)

#Choose correct columns
vacc_den = vacc_den.iloc[: , [0, 1, 2, -1]]

#Strip %
vacc_den['percentage_finished vaccinated'] = vacc_den['percentage_finished vaccinated'].str.replace(r'%', '')


#Group vaccination dataset
vacc_den = vacc_den.groupby(['municipality', 'percentage_finished vaccinated', 'ppl_inventory', 'parish'], as_index=False).sum();

#Change datatype of %vaccinated
vacc_den["percentage_finished vaccinated"] = vacc_den["percentage_finished vaccinated"].astype(float, 'raise')


#Calculate the amount of vaccinated people per parish
vacc_den['tot_vacc_parish'] = vacc_den['ppl_inventory'] * (vacc_den['percentage_finished vaccinated'] / 100);

#Group the dataset into municipalities 
vacc_den = vacc_den.groupby(['municipality'], as_index=False).sum(['ppl_inventory', 'tot_vacc_parish']);
vacc_den = vacc_den.iloc[: , [0, 2, 3]]

#Calculate the percentage of infected by municipality 
vacc_den['vacc_pct_municipality'] = vacc_den['tot_vacc_parish'] / vacc_den['ppl_inventory'] * 100

vacc_den = vacc_den.rename(columns={"tot_vacc_parish" : "tot_vacc_municipality"})


print(vacc_den)
```
